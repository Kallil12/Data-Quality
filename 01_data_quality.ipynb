# For exercise 3 we must do some data cleaning.


```python
# 3A

df_retail['RETAIL'] = 1
```

```python
# 3B

df_dealer['CUSTOMER_NUMBER'] = df_dealer['CUSTOMER_NUMBER'].astype(int)
```

```python
# 3B

df_retail['REG_DEALER_NUMBER'] = df_retail['REG_DEALER_NUMBER'].astype(int)
```

```python
# 3C

region_mapping = {
    1: ["AB", "BC", "MB", "NB", "NF", "NS", "NT", "NU", "ON", "PE", "QC", "SK", "YT"],
    2: ["AL", "CT", "DE", "DC", "FL", "GA", "MA", "MD", "ME", "NC", "NH", "NJ", "NY", "PA", "PR", "RI", "SC", "VA", "VT", "WV"],
    3: ["AR", "IA", "IL", "IN", "KS", "KY", "LA", "MI", "MN", "MO", "MS", "ND", "NE", "OH", "SD", "TN", "WI"],
    4: ["AK", "AZ", "CA", "CO", "HI", "ID", "MT", "NM", "NV", "OK", "OR", "TX", "UT", "WA", "WY"]
}

df_dealer['REGION_CODE'] = df_dealer['STATE_CODE'].apply(
    lambda state: next((region for region, states in region_mapping.items() if state in states), 5)
)
```

# For exercise 4 we must do some high level analysis

```python
# 4A

df_merged_data = pd.merge(df_retail, df_dealer, how='outer', left_on='REG_DEALER_NUMBER', right_on='CUSTOMER_NUMBER', indicator=True)

# Create the 'CHECK_FLAG' column based on the indicator
df_merged_data['CHECK_FLAG'] = 'OK'
df_merged_data['CHECK_FLAG'][df_merged_data['_merge'] == 'left_only'] = 'CHECK DATA RETAIL'
df_merged_data['CHECK_FLAG'][df_merged_data['_merge'] == 'right_only'] = 'CHECK DATA DEALER'

# Drop the '_merge' column if you don't need it
df_merged_data.drop(columns=['_merge'], inplace=True)
```

```python
df_merged_data
```

```python
# 4B I

df_merged_data_filtered = df_merged_data[df_merged_data['CHECK_FLAG'] == 'OK']

df_az_mb_sales = df_merged_data_filtered[(df_merged_data_filtered['STATE_CODE'].isin(['AZ', 'MB']))]
df_top_selling_model_az_mb = df_az_mb_sales.groupby('MODEL_CODE')['RETAIL'].sum().idxmax()
```
